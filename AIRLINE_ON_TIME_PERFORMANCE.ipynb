```python
df_filepath="/content/drive/MyDrive/LSN TSAP/AIRLINE PERFORMANCE.csv"
df=pd.read_csv(df_filepath, index_col="MONTH")
df.head()
```

```python
flights = sns.load_dataset('flights')
sns.lineplot(data=flights,x='month',y='passengers')
```

```python
plt.title('Monthly Visualization: \nTotal number of flights: Green,  \nNumber of on time flights: Blue, \nNumber of delayed flights: Red')
sns.lineplot(data=df, x='MONTH',y='NO OF FLIGHTS',color='green');
sns.lineplot(data=df, x='MONTH',y='NO OF ON TIME FLIGHTS',color='blue');
sns.lineplot(data=df, x='MONTH',y='NO OF DELAYED FLIGHTS',color='red');
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["JAN", "FEB", "MAR", "APRIL"]

plt.pie(y, labels = mylabels)
plt.title("Visualization of Cancelled Flights for first quarter of the year")
plt.legend(title ="Cancelled flights:")
plt.show()
```

```python
# Plot the heatmap
col_list = ['NO OF FLIGHTS', 'NO OF PASSENGERS', 'NO OF DELAYED FLIGHTS']
plt.figure(figsize=(15, 7))
sns.heatmap(df[col_list].corr(), annot=True, vmin=-1, vmax=1, fmt=".2f", cmap="Spectral")
plt.title("HOW NO OF FLIGHTS AND NO OF PASSENGERS CONTRIBUTE TO NO OF DELAYS?")
plt.show()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8, 6))
# Bar chart showing how no of passengers influence delays
sns.barplot(x=df['NO OF DELAYED FLIGHTS'], y=df['NO OF PASSENGERS'])
# Add label for horizontal axis
plt.xlabel("NO OF DELAYED FLTS")
# Add label for vertical axis
plt.title("HOW NUMBER OF PASSENGERS INFLUENCE DELAYS");
```
